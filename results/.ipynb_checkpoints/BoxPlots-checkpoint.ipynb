```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
for counter in range(1,7):
    df=pd.read_csv('output'+str(counter)+'.csv')
    df_sense=df[df['totalStockRatio']==.25]
    df_sense=df_sense[df_sense['recoveryRate']==1]
    df_sense=df_sense[df_sense['pooledFraction']==.5]
    
    df_sense_pool=df_sense[df_sense['sensitivityType']=='pool']
    df_sense_safe=df_sense[df_sense['sensitivityType']=='safety']
    pool1=df_sense_pool[df_sense_pool['sensitivityParameter']==.1]
    pool2=df_sense_pool[df_sense_pool['sensitivityParameter']==.2]
    safe1=df_sense_safe[df_sense_safe['sensitivityParameter']==.1]
    safe2=df_sense_safe[df_sense_safe['sensitivityParameter']==.2]
    base=df_sense[df_sense['sensitivityType']=='base']

    fig, ax = plt.subplots()
    medianlineprops = dict(color='black')
    meanpointprops = dict(marker='D', markeredgecolor='black',markersize=1.5,
                          markerfacecolor='black')
    flierprops = dict(marker='o', markersize=1,
                      linestyle='none')
    bp1=ax.boxplot(np.array(base['Type I']),positions = [1],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    bp2=ax.boxplot(np.array(base['Type II']),positions = [2],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(pool1['Type I']),positions = [4],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(pool1['Type II']),positions = [5],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(pool2['Type I']),positions = [7],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(pool2['Type II']),positions = [8],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(safe1['Type I']),positions = [10],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(safe1['Type II']),positions = [11],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(safe2['Type I']),positions = [13],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
    ax.boxplot(np.array(safe2['Type II']),positions = [14],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)

    plt.axhline(base['Type I'].mean(), c='orange')
    plt.axhline(base['Type II'].mean(), c='green')
    plt.axhline(base['Type I'].median(), c='orange', linestyle=':')
    plt.axhline(base['Type II'].median(), c='green', linestyle=':')

    ax.legend([bp1["boxes"][0], bp2["boxes"][0]], ['Type I', 'Type II'], loc='upper right')

    ax.set_xticks([1.5,4.5,7.5,10.5,13.5])
    ax.set_xticklabels(['Optimal','10% Pooled','20% Pooled','10% Safety','20% Safety'])

    plt.savefig('boxcompare_'+str(counter)+'.png',dpi=600)

    plt.show()
```

```python
#POOLED

for counter in range(1,7):
    df=pd.read_csv('output'+str(counter)+'.csv')
    df=df[df['sensitivityType']=='base']

    df1=df[df['totalStockRatio']==.25]
    df1=df1[df1['recoveryRate']==1]
        
    
    fig, ax = plt.subplots()
    medianlineprops = dict(color='black')
    meanpointprops = dict(marker='D', markeredgecolor='black',markersize=1.5,
                          markerfacecolor='black')
    flierprops = dict(marker='o', markersize=1,
                      linestyle='none')
    
    ticklabels=[]
    for i in range(5):
        pooled=i*0.25
        ticklabels=ticklabels+[str(pooled*100)+'%']
        bp1=ax.boxplot(np.array(df1[df1['pooledFraction']==pooled]['Type I']),positions = [3*i+1],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
        bp2=ax.boxplot(np.array(df1[df1['pooledFraction']==pooled]['Type II']),positions = [3*i+2],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
        
        if i==0:
            plt.axhline(df1[df1['pooledFraction']==pooled]['Type I'].mean(), c='orange')
            plt.axhline(df1[df1['pooledFraction']==pooled]['Type II'].mean(), c='green')
            plt.axhline(df1[df1['pooledFraction']==pooled]['Type I'].median(), c='orange', linestyle=':')
            plt.axhline(df1[df1['pooledFraction']==pooled]['Type II'].median(), c='green', linestyle=':')

    ax.legend([bp1["boxes"][0], bp2["boxes"][0]], ['Type I', 'Type II'], loc='upper right')

    ax.set_xticks([1.5,4.5,7.5,10.5,13.5])
    ax.set_xticklabels(ticklabels)

    plt.savefig('pooled_'+str(counter)+'.png',dpi=600)

    plt.show()
```

```python
#total stock

for counter in range(1,7):
    df=pd.read_csv('output'+str(counter)+'.csv')
    df=df[df['sensitivityType']=='base']

    df1=df[df['recoveryRate']==1]
    df1=df1[df1['pooledFraction']==.5]
    
    fig, ax = plt.subplots()
    medianlineprops = dict(color='black')
    meanpointprops = dict(marker='D', markeredgecolor='black',markersize=1.5,
                          markerfacecolor='black')
    flierprops = dict(marker='o', markersize=1,
                      linestyle='none')
    
    ticklabels=[]
    for i in range(1,5):
        totStock=i*0.25
        ticklabels=ticklabels+[str(totStock*100)+'%']
        bp1=ax.boxplot(np.array(df1[df1['totalStockRatio']==totStock]['Type I']),positions = [3*(i-1)+1],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
        bp2=ax.boxplot(np.array(df1[df1['totalStockRatio']==totStock]['Type II']),positions = [3*(i-1)+2],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
        
        if i==0:
            plt.axhline(df1[df1['totalStockRatio']==totStock]['Type I'].mean(), c='orange')
            plt.axhline(df1[df1['totalStockRatio']==totStock]['Type II'].mean(), c='green')
            plt.axhline(df1[df1['totalStockRatio']==totStock]['Type I'].median(), c='orange', linestyle=':')
            plt.axhline(df1[df1['totalStockRatio']==totStock]['Type II'].median(), c='green', linestyle=':')

    ax.legend([bp1["boxes"][0], bp2["boxes"][0]], ['Type I', 'Type II'], loc='upper right')

    ax.set_xticks([1.5,4.5,7.5,10.5])
    ax.set_xticklabels(ticklabels)

    plt.savefig('totalStock_'+str(counter)+'.png',dpi=600)

    plt.show()
```

```python
#recovery

for counter in range(1,7):
    df=pd.read_csv('output'+str(counter)+'.csv')
    df=df[df['sensitivityType']=='base']
    
    df1=df[df['totalStockRatio']==.25]
    df1=df1[df1['pooledFraction']==.5]
    
    fig, ax = plt.subplots()
    medianlineprops = dict(color='black')
    meanpointprops = dict(marker='D', markeredgecolor='black',markersize=1.5,
                          markerfacecolor='black')
    flierprops = dict(marker='o', markersize=1,
                      linestyle='none')
    
    ticklabels=[]
    for i in range(1,7):
        recovery=i
        ticklabels=ticklabels+[str(recovery)]
        bp1=ax.boxplot(np.array(df1[df1['recoveryRate']==recovery]['Type I']),positions = [3*i+1],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C1"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
        bp2=ax.boxplot(np.array(df1[df1['recoveryRate']==recovery]['Type II']),positions = [3*i+2],notch=True,flierprops=flierprops, patch_artist=True,boxprops=dict(facecolor="C2"),medianprops=medianlineprops,meanprops=meanpointprops,showmeans=True, widths = 0.5)
        
        if i==0:
            plt.axhline(df1[df1['recoveryRate']==recovery]['Type I'].mean(), c='orange')
            plt.axhline(df1[df1['recoveryRate']==recovery]['Type II'].mean(), c='green')
            plt.axhline(df1[df1['recoveryRate']==recovery]['Type I'].median(), c='orange', linestyle=':')
            plt.axhline(df1[df1['recoveryRate']==recovery]['Type II'].median(), c='green', linestyle=':')

    ax.legend([bp1["boxes"][0], bp2["boxes"][0]], ['Type I', 'Type II'], loc='upper right')

    ax.set_xticks([1.5,4.5,7.5,10.5,13.5,16.5])
    ax.set_xticklabels(ticklabels)

    plt.savefig('recovery_'+str(counter)+'.png',dpi=600)

    plt.show()
```
